```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
def plot_elbow(data, max_k=10):
    wcss = []
    for k in range(1, max_k + 1):
        km = KMeans(n_clusters=k, init='k-means++', 
                    random_state=42, n_init=10)
        km.fit(data)
        wcss.append(km.inertia_)
    
    plt.figure(figsize=(8, 5))
    plt.plot(range(1, max_k + 1), wcss, marker='o', linestyle='--')
    plt.title('Elbow Method')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('WCSS (Inertia)')
    plt.show()
    return wcss
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df[['sepal_length', 'sepal_width']]
X.head()

```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
k = 3
```

```python
km = KMeans(
    n_clusters=k, 
    init='k-means++',
    max_iter=300, 
    random_state=42
)

km.fit(X_scaled)
```

```python
labels = km.labels_
labels[:75]
```

```python
df['cluster'] = labels
df.head(60)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=df['sepal_length'], 
    y=df['sepal_width'], 
    hue=df['cluster'], 
    palette='Set1'
)

plt.title('K-means Clustering of Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Cluster')
plt.show()
```

```python
# finding optimal k using elbow method
wcss = plot_elbow(X_scaled, max_k=10)
```

```python
intertia = []

k_range = range(1, 11)

for k in k_range:
    km = KMeans(
        n_clusters=k, 
        init='k-means++', 
        random_state=42,
    )
    
    km.fit(X_scaled)
    intertia.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, intertia, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Inertia)')
plt.show()
```

```python
score = silhouette_score(X_scaled, labels)
print(f'Silhouette Score: {score:.2f}')
```

Interpretation of Score: 
- Above 0.5: Strong clustering structure
- Between 0.25 and 0.5: Overlapping clusters
- Near 0: Useless clustering structure

```python
k_vals = range(2, 11)

for k in k_vals:
    km = KMeans(
        n_clusters=int(k), 
        init='k-means++', 
        random_state=42,
        n_init=10
    )
    
    km.fit(X_scaled)
    score = silhouette_score(X_scaled, km.labels_)
    print(f'K: {k:.2f}, Silhouette Score: {score:.2f}')
```
