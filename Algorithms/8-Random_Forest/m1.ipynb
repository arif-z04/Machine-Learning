# What is Random Forest? 

- Ensemble learning Algorithm that combines multiple decision trees to improve predictive performance and reduce overfitting. 
- Multiple Decision Trees: Built from many independent decision trees working together as a unified system. 
- Dual Purpose: Handles both classification and regression tasks with equal effectiveness.

# Why Random Forest was developed?

Single Decision trees, while intuitive and easy to interprest, face significant challenges in real-world applications:
- Overfitting: Decision trees can easily overfit the training data, leading to poor generalization on unseen data.
- Small Changes in training data can lead to different tree structures, making them unstable and less reliable.
- Predictions often lack stability and reliability. 

**The solution: Random Forest** : Random forest was created to address these fundamental weaknesses by harnessing the wisdom of crowds. By building multiple decision trees and aggregating their predictions, random forest significantly enhances the robustness and accuracy of the model, making it a powerful tool for both classification and regression tasks.


# Core Idea of Random Forest
- **Train Many Trees:** Create hundreds or thousands of decision trees, each trained on a different random subset of the training data (bootstrap sampling).
- **Add diversity:** Each tree sees a random subset of data and features, ensuring unique perspectives and reducing correlation between trees.
- **Aggregate Predictions:** For classification, the final prediction is made by majority voting among the trees; for regression, it is the average of the predictions from all trees.

# How Random Forest Works: 



# What Random Forest Improves: 
- **Reduces Overfitting:** By averaging the predictions of multiple trees, random forest reduces the risk of overfitting that single decision trees often face.
- **Improves Accuracy:** The ensemble approach typically leads to higher accuracy and better generalization on unseen data compared to individual decision trees.
- **Stable Predictions:** The aggregated predictions from multiple trees result in more stable and reliable outcomes compared to individual decision trees.

# Where random forest is used? 

- **Classification Tasks:** Random forest is widely used for classification problems, such as spam detection, image recognition, and medical diagnosis.
- **Complex Regression Tasks:** It is also effective for regression problems, such as predicting house prices, stock prices, and customer lifetime value.
- **Feature Importance:** Random forest can provide insights into feature importance, helping identify which features are most influential in making predictions.
- **Handling Missing Data:** Random forest can handle missing data effectively, making it a robust choice for real-world datasets that may have incomplete information.
- **Tabular Datesets** : Random forest is particularly well-suited for tabular datasets with structured data, where it can capture complex relationships between features and target variables.

# What is a Decision Tree?
- A decision tree is a tree-structured model that uses a series of decision rules to make predictions based on input features. At each internal node, the tree evaluates a feature and splits the data based on condition. These splits continue until reaching leaf nodes, where the final out or prediction is produced. These splits continue until reaching leaf nodes, where the final output or prediction is produced. Think of it as a flowchart that guides you through a series of yes/no questions to arrive at a decision or prediction.


# How decision Trees Make Decisions?

## The Splitting Process:
1. **Evaluate All Features:** Consider every possible split point fore ach feature in the dataset.
2. **Measure Purity Gain:** Calculate how well each potential split separates the data into distinct classes (for classification) or reduces variance (for regression).
3. **Choose Best Split** : Select the feature and split point that provides the highest purity gain or variance reduction.
4. **Repeat Recursively:** Apply the same splitting process to each child node until a stopping criterion is met (e.g., maximum depth, minimum samples per leaf, or pure nodes).

## Limitations of Decision Trees:
- **Overfitting:** Decision trees can easily overfit the training data, especially if they are allowed to grow deep without constraints. This leads to poor generalization on unseen data.
- **Instability:** Small changes in the training data can lead to different tree structures, making them unstable and less reliable.
- **Bias towards Dominant Features:** Decision trees may favor features with more levels or unique values, which can lead to biased splits and suboptimal performance. 
- **Learning Noise:** Decision trees can learn the noise in the training data, leading to poor performance on unseen data. 
- **High variance:** small changes in the training data can lead to a completely different tree structure, making them unstable and less reliable. 

## Bias and Variance: 
- **Bias:** Refers to the error introduced by approximating a real-world problem with a simplified model. High bias can lead to underfitting, where the model fails to capture the underlying patterns in the data.
- **Variance:** Refers to the error introduced by the model's sensitivity to fluctuations in the training data. High variance can lead to overfitting, where the model captures noise in the training data rather than the true underlying patterns.
- **Trade-off:** Decision trees often have low bias but high variance, making them prone to overfitting. Random forest addresses this issue by averaging the predictions of multiple trees, which reduces variance and improves generalization.


## Ensemble Learning Methods: 
- **Bagging Approach:** Random forest uses a bagging (bootstrap aggregating) approach, where multiple decision trees are trained on different random subsets of the training data. This helps to reduce variance and improve generalization.



- **Boosting Approach:** Boosting is another ensemble learning method that focuses on sequentially training models, where each model attempts to correct the errors of the previous one. Unlike bagging, boosting gives more weight to misclassified instances, allowing the model to learn from its mistakes and improve performance over time. Examples of boosting algorithms include AdaBoost, Gradient Boosting, and XGBoost. 



# How ensemble Combines Models: 

**Classification Tasks:**
for classification problems, ensembles use __*majority voting*__ to combine the predictions of individual models. Each model in the ensemble makes a prediction, and the final output is the class that receives the most votes from the models.

**Regression Tasks:**
for regression problems, ensembles typically use __*averaging*__ to combine the predictions. Each model in the ensemble produces a numerical prediction, and the final output is the average of these predictions.

1. **Multiple Base Models**: An ensemble consists of multiple base models, such as decision trees, that are trained independently on different subsets of the training data.
2. **Individual Predictions**: Each base model makes its own prediction based on the input features.
3. **Aggregation**: The predictions from all base models are aggregated to produce a final prediction.
4. **Final Prediction**: For classification, the final prediction is the class that receives the most votes from the base models. For regression, it is the average of the predictions from all base models.

# Bootstrap Sampling: 
- **Definition**: Bootstrap sampling is a technique used in ensemble learning where multiple subsets of the training data are created by randomly sampling with replacement. Each subset is used to train a different base model in the ensemble.
- **Purpose**: The main purpose of bootstrap sampling is to introduce diversity among the base models, which helps to reduce variance and improve the generalization of the ensemble. By training each model on a different subset of the data, the ensemble can capture different patterns and relationships, leading to more robust predictions.
- **Process**: In bootstrap sampling, each subset is created by randomly selecting samples from the original training data, allowing for duplicates. This means that some samples may be included multiple times in a subset, while others may not be included at all. This randomness helps to create diverse models that can capture different aspects of the data, ultimately improving the performance of the ensemble. 
- **Example**: If you have a training dataset with 100 samples, you can create multiple bootstrap samples by randomly selecting 100 samples with replacement. Each bootstrap sample will contain some duplicates and some unique samples, providing different perspectives for training the base models in the ensemble.
  
 



# Feature Randomness: 
- **Definition**: Feature randomness is a technique used in ensemble learning, particularly in random forest, where a random subset of features is selected for each split in the decision trees. This means that at each node of the tree, only a random subset of features is considered for splitting the data.
- **Purpose**: The main purpose of feature randomness is to introduce additional diversity among the trees in the ensemble. By considering only a random subset of features at each split, the trees are forced to explore different feature combinations and relationships, which helps to reduce correlation between the trees and improve the overall performance of the ensemble.
- **Process**: In random forest, when building each decision tree, a random subset of features is selected at each split. This means that instead of considering all features for splitting the data, only a random subset is evaluated. This randomness helps to create diverse trees that can capture different patterns in the data, leading to improved generalization and reduced overfitting.
- **Example**: If you have a dataset with 10 features, you can specify that at each split in the decision tree, only 3 randomly selected features are considered for splitting the data. This means that each tree in the random forest will be built using different combinations of features, which helps to create a more robust and accurate ensemble model.

## Feature Randomness in Each Split:

1. **Full Feature Set Available**
2. **Random Subset of Features**
3. **Best Split from Random Subset** 

Example: 
Imagine a credit risk model with features: Age, Income, Debt, Credit History, and Employment Status. 
- Split 1: Random subset: Income, Debt: Chosen split: Income > $50,000
- Split 2: Random subset: Age, Credit History: Chosen split: Age < 30
- Split 3: Random subset: Employment Status, Income: Chosen split: Income > $70,000

# Implementation of Random Forest: 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, 
    confusion_matrix,
    classification_report
)
```

```python
data = load_breast_cancer()
X = data.data
y = data.target

X = pd.DataFrame(X, columns=data.feature_names)
y = pd.Series(y, name='target')

print("Feature matrix shape: ", X.shape)
print("Target shape: ", y.shape)

```

```python
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y,
)

print("Training samples: ", X_train.shape[0])
print("Test samples: ", X_test.shape[0])
```

```python
rf_baseline = RandomForestClassifier(
    n_estimators=100, # how many trees should be there
    random_state=42
) 

rf_baseline.fit(X_train, y_train) 
```

```python
y_pred = rf_baseline.predict(X_test)
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
print("Classification Report: \n", classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=data.target_names,
            yticklabels=data.target_names
            )
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()  
```

```python
# Feauture importance
importances = pd.Series(
    rf_baseline.feature_importances_, 
    index=X.columns
).sort_values(ascending=False)

importances.head(10)
```

```python
plt.figure(figsize=(8,6))
importances.head(10).plot(kind='bar')
plt.title('Top 10 Feature Importances')
plt.ylabel('Importance Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Why Hyperparameter Tuning is important:

Problems with default hyperparameters:
- Trees maybe too deep
- Model may overfit
- Training may be uncessarily slow

Key Hyperparameters to Tune:
- `n_estimators`: Number of trees in the forest. More trees can improve performance but also increase training time.
- `max_depth`: Maximum depth of each tree. Limiting depth can help prevent overfitting.
- `min_samples_split`: Minimum number of samples required to split an internal node. Higher values can lead to simpler trees and reduce overfitting.
- `max_features`: Maximum number of features to consider when looking for the best split. Limiting features can help prevent overfitting.

```python
# key Hyperparameters 

param_grid = {
    'n_estimators': [100, 150, 175, 200],
    'max_depth': [None,5, 10, 15],
    'min_samples_split': [2, 3, 5],
    'max_features': ['sqrt', 'log2']
}

# Grid SearchCV for hyperparameter tuning on cpu
grid_rf = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1
)

grid_rf.fit(X_train, y_train)
```

```python
print("Best parameters found: ")
print(grid_rf.best_params_)
```

```python
best_rf = grid_rf.best_estimator_
y_pred_best = best_rf.predict(X_test)

print("Accuracy: ", accuracy_score(y_test, y_pred_best))
print("Classification Report: \n", classification_report(y_test, y_pred_best))

cm = confusion_matrix(y_test, y_pred_best)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=data.target_names,
            yticklabels=data.target_names
            )
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Tuned RF')
plt.show()
```
