# Naive Bayes Classifier

## Bayes Theorem কী?

Bayes Theorem হলো probability (সম্ভাবনা) বের করার একটি সূত্র।
এটা ব্যবহার করা হয় কোনো নতুন তথ্য পাওয়ার পরে কোনো ঘটনার সম্ভাবনা নির্ণয় করতে।

$$
\mathbf{P(A|B) = \frac{P(B|A)P(A)}{P(B)}} \quad \text{(Bayes' Theorem)}
$$

সহজভাবে:

> “কোনো ঘটনা ঘটেছে জেনে আরেকটি ঘটনার সম্ভাবনা কত” — এটা বের করতে Bayes Theorem ব্যবহার করা হয়।

### উদাহরণ:

ধরো,

* কারো রোগ থাকার সম্ভাবনা ১%
* টেস্ট ৯৯% সঠিক

এখন যদি রিপোর্ট Positive আসে,
তাহলে আসলে তার রোগ থাকার সম্ভাবনা কত — এটা Bayes theorem দিয়ে বের করা যায়।

## “Flipping Bayes Rule” কী?

“Flipping Bayes Rule” বলতে বোঝায়:

> কোনো probability কে উল্টে ফেলা বা reverse করা।

মানে:

আমরা অনেক সময় জানি:

* $(P(B|A))$ → “A হলে B হওয়ার সম্ভাবনা”

কিন্তু জানতে চাই:

* $(P(A|B))$ → “B হলে A হওয়ার সম্ভাবনা”

এই উল্টানো কাজটাই Bayes Theorem করে।

---

উদাহরণ:

ধরো,

* যদি বৃষ্টি হয়, রাস্তা ভিজে যায়।

মানে:

* $(P(\text{Wet Road} | \text{Rain}))$

কিন্তু তুমি জানতে চাও:

> রাস্তা ভেজা দেখে বৃষ্টি হওয়ার সম্ভাবনা কত?

মানে:

* $(P(\text{Rain} | \text{Wet Road}))$

এখানে probability উল্টানো হচ্ছে।
এই “reverse” বা “flip” করাকেই বলা হয়:

> Flipping Bayes Rule

---

সহজ ভাষায়:

Bayes Theorem ব্যবহার হয়:

> “কারণ থেকে ফল” জানা থাকলে
> “ফল দেখে কারণ” অনুমান করার জন্য।


# Bayes Theorem POV from code 

$$
P(Y|X) = \frac{P(X|Y)P(Y)}{P(X)}
$$ 

এখানে: 
- $X$ = Data - (features)(যা আমরা দেখি) 
- $Y$ = Class (যা আমরা predict করতে চাই)
- $P(Y)$ = Prior probability (Class এর আগের সম্ভাবনা)
- $P(X|Y)$ = Likelihood (Class এর জন্য data এর সম্ভাবনা)
- $P(X)$ = Evidence (Data এর মোট সম্ভাবনা) 
- $P(Y|X)$ = Posterior probability (Data দেখে Class এর সম্ভাবনা) 

# Simple Example:

Assume: 
- 40 percent of emails are spam.
- 20 percent of spam emails contain the word "lottery".
- 2 percent of emails are spam and contain the word "lottery".

Suppose you receive an email that contains the word
"lottery". What is the probability that this email is spam? 
calculate $P(Spam|Lottery)$

Let, 
- Spam = S
- Lottery = L
- Not Spam = NS
  
Given:
- $P(S) = 0.4$
- $P(L|S) = 0.2$
- $P(L|NS) = 0.02$ 

To calculate $P(S|L)$ using Bayes Theorem, we need to find $P(L)$ first.

$$
P(L) = P(L|S)P(S) + P(L|NS)P(NS)
$$

$$
P(L) = (0.2)(0.4) + (0.02)(0.6) = 0.08 + 0.012 = 0.092
$$

Now, we can calculate $P(S|L)$:

$$
P(S|L) = \frac{P(L|S)P(S)}{P(L)} = \frac{(0.2)(0.4)}{0.092} = \frac{0.08}{0.092} \approx 0.8696
$$


# The Naive Assumption

For features $X_1, X_2, ..., X_n$ and class $Y$, the Naive Bayes classifier assumes that:

$$
P(X_1, X_2, ..., X_n | Y) = \sum_{i=1}^{n} P(X_i | Y)
$$

Example: 
Suppose we have two features:
- $X_1$ = "Contains the word 'offer'"
- $X_2$ = "Contains the word 'free'"
- $X_3$ = "Contains the word 'win'"
- $Y$ = "Spam"

The Naive Bayes classifier would calculate:
$$
P(X_1, X_2, X_3 | Y) = P(X_1 | Y) + P(X_2 | Y) + P(X_3 | Y)
$$

# Training and Prediction Mechanics: 

## 1. Goal of Naive Bayes:
   
- To find the class $Y$ that maximizes the posterior probability given the data $X$.
- This can be expressed as:
   
$$  \hat{Y} = argmax P(Y|X) = argmax P(X|Y)P(Y)$$

## 2. Training Phase:

**a. Calculate Prior Probabilities:**   

$$ P(Y) = C_k $$
   where $C_k$ is the count of samples in class $k$ divided by the total number of samples.

**b. Calculate Likelihoods:**    

$$ P(X_i | Y) = \frac{Count(X_i, Y)}{Count(Y)} $$

where $Count(X_i, Y)$ is the count of samples in class $Y$ that have feature $X_i$, and $Count(Y)$ is the total count of samples in class $Y$.

**Laplace Smoothing:**

To handle zero probabilities, Laplace smoothing is often applied:

$$ P(X_i | Y) = \frac{Count(X_i, Y) + 1}{Count(Y) + V} $$

where $V$ is the number of possible values for feature $X_i$.


## 3. Prediction Phase:

To predict the class for a new data point, we calculate the posterior probability for each class and select the class with the highest probability:
$$  \hat{Y} = argmax P(Y|X) = argmax P(X|Y)P(Y) $$



# Types of Naive Bayes Classifiers:

1. **Multinomial Naive Bayes:** Used for discrete data, especially in text classification where features represent word counts or frequencies.
2. **Bernoulli Naive Bayes:** Used for binary/boolean features, where features represent the presence or absence of a word.
3. **Gaussian Naive Bayes:** Used for continuous data, where features are assumed to follow a normal distribution.
4. **Complement Naive Bayes:** A variant of Multinomial Naive Bayes that is particularly effective for imbalanced datasets. It focuses on the complement of each class to improve performance.
5. **Categorical Naive Bayes:** Used for categorical data, where features represent categories rather than counts or continuous values.






# Gaussian Naive Bayes on breast cancer dataset:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_diabetes, load_breast_cancer, load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, classification_report, confusion_matrix, accuracy_score
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
data = load_breast_cancer()
X = data.data
y = data.target

print("Shape of X (features):", X.shape)
print("Shape of y (target):", y.shape) 
print("Target names:", data.target_names)
print("First 5 feature names:\n")
print(data.feature_names[:5])
```

```python
df = pd.DataFrame(X, columns=data.feature_names)
df["target"] = y
df.head()
```

```python
# Class distribution

class_counts = df["target"].value_counts()
print(class_counts)

plt.figure(figsize=(6, 4))
class_counts.plot(kind='bar')
plt.xticks(ticks=[0,1], labels=data.target_names, rotation=0)
plt.title("Class distribution (0 = malignant, 1 = benign)")
plt.ylabel("Count")
plt.show()
```

```python
# Visualize all feature distributions
# feature_name = "mean radius"
# feat_idx = list(data.feature_names).index(feature_name)

# plt.figure()
# plt.hist(X[y==0][:, feat_idx], bins=30, alpha=0.5, label="malignant (0)")
# plt.hist(X[y==1][:, feat_idx], bins=30, alpha=0.5, label="benign (1)")
# plt.xlabel(feature_name)
# plt.ylabel("Frequency")
# plt.title("Distribution of {}".format(feature_name))
# plt.legend()
# plt.show()


for feature in data.feature_names:
    # plot all distribution into single page
    plt.figure(figsize=(6, 4))
    plt.hist(X[y==0][:, list(data.feature_names).index(feature)], bins=30, alpha=0.5, label="malignant (0)")
    plt.hist(X[y==1][:, list(data.feature_names).index(feature)], bins=30, alpha=0.5, label="benign (1)")
    plt.title(f"Distribution of \"{feature}\" by target class")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.legend(title="Target", labels=data.target_names)
    plt.show()
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print("Training set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)
```

```python
# Train GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
# Predict and evaluate

y_pred = gnb.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print("GaussianNB Accuracy:", acc)

print("Classification Report:\n", classification_report(y_test, y_pred, target_names=data.target_names))
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=data.target_names, yticklabels=data.target_names)

plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.title("GaussianNB - Confusion Matrix")
plt.show()
```

# Multinomial Naive Bayes:

```python
texts = [
    "I love this product",
    "This is the worst experience ever",
    "Absolutely fantastic service",
    "I will never buy this again",
    "I hate this item",
    "I love this movie",
    "I don't like this",
    "Horrible and Terrible Experience"
]

labels = [1, 0, 1, 0, 0, 1, 0, 0]  # 1 = positive, 0 = negative
toy_df = pd.DataFrame({"text": texts, "label": labels})
print(toy_df.head())
```

```python
# Convert text into Features with CountVectorizer
vectorizer = CountVectorizer()
X_toy = vectorizer.fit_transform(toy_df['text'])
y_toy = toy_df['label']

print("Feature matrix shape:", X_toy.shape)

print("Vocuabulary:", vectorizer.get_feature_names_out())
```

```python
mnb_toy = MultinomialNB()
mnb_toy.fit(X_toy, y_toy)

y_toy_pred = mnb_toy.predict(X_toy)
acc_toy = accuracy_score(y_toy, y_toy_pred)

print("MultinomialNB Accuracy on toy dataset:", acc_toy)
print("classification_report:\n", classification_report(y_toy, y_toy_pred))
```

```python
new_texts = [
    "I really love this product",
    "This is the worst thing I have ever bought",
    "Fantastic experience, highly recommend",
    "I will never buy this again, terrible",
    "I hate this item, it's awful",
    "I love this movie, it's amazing",
    "I don't like this at all",
    "Horrible and Terrible Experience, do not buy"
]

new_labels = [1, 0, 1, 0, 0, 1, 0, 0]  # For reference, not used in prediction

X_new = vectorizer.transform(new_texts)
y_new_pred = mnb_toy.predict(X_new)

for text, pred in zip(new_texts, y_new_pred):
    label = "Positive" if pred == 1 else "Negative"
    print(f"Text: {text!r} --> Predicted Sentiment: {label}")
```

# Multinomial Naive Bayes on 20 Newsgroups dataset:

```python
from sklearn.datasets import fetch_20newsgroups

```

```python
categories = ['comp.graphics', 'rec.sport.baseball', 'sci.med']

newsgroups = fetch_20newsgroups(
    subset='train', 
    categories=categories, 
    remove=('headers', 'footers', 'quotes'),
    shuffle=True,
    random_state=42    
)
```

```python
df_news = pd.DataFrame({
    "text": newsgroups.data,
    "label": newsgroups.target
})

df_news.head()
```

```python
print("Number of documents:", len(newsgroups.data))
print("Target names:", newsgroups.target_names)
```

```python
# train test split

X_train_text, X_test_text, y_train_text, y_test_text = train_test_split(
    df_news['text'], df_news['label'], test_size=0.25, random_state=42
)


print("Train Size: ", X_train_text.shape[0])
print("Test Size: ", X_test_text.shape[0])
```

```python
vectorizer = CountVectorizer(stop_words='english', max_features=4000)

X_train_count = vectorizer.fit_transform(X_train_text)
X_test_count = vectorizer.transform(X_test_text)

print("Shape of X_train_count:", X_train_count.shape)
print("Shape of X_test_count:", X_test_count.shape)
```

```python
mnb_news = MultinomialNB()
mnb_news.fit(X_train_count, y_train_text)

print("MultinomialNB trained on 20 Newsgroups dataset.")
```

```python
y_new_pred = mnb_news.predict(X_test_count)
acc_news = accuracy_score(y_test_text, y_new_pred) 

print("Accuracy on 20 Newsgroups test set:", acc_news * 100)

cm_news = confusion_matrix(y_test_text, y_new_pred)
print("Confusion Matrix:\n", cm_news)

sns.heatmap(cm_news, annot=True, fmt='d', cmap='Greens',
            xticklabels=newsgroups.target_names, 
            yticklabels=newsgroups.target_names)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("MultinomialNB - Confusion Matrix (20 Newsgroups)")
plt.show()

print("Classification Report:\n", classification_report(y_test_text, y_new_pred, target_names=newsgroups.target_names))
```
