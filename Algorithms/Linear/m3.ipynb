# Linear Regression: Assumptions & Limitations

## 1. Why Assumptions Matter
- Unreliable Predictions: Violating assumptions can lead to inaccurate predictions.
- Misleading coefficients: Coefficients may not reflect true relationships.
- Increased error: Violations can increase prediction errors.
- Invalid inference: Statistical tests may become unreliable.
- Model instability: Small changes in data can lead to large changes in coefficients.
- Poor generalization: Model may not perform well on new data.
