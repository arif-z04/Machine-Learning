```python
from sklearn.linear_model import LinearRegression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

def load_data(file_path):
    if os.path.exists(file_path):
        data = pd.read_csv(file_path)
        return data
    else:        
        print(f"File {file_path} not found.")
        return None
    
```

```python
df = load_data("student_study_hours_gpa.csv")
```

```python
hours = np.array(df["StudyHours"]).reshape(-1, 1)
gpa = np.array(df["GPA"]).reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(hours, gpa)
```

```python
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```

```python
plt.figure(figsize=(10, 6)) 
plt.scatter(hours, gpa, color='blue', label='Data Points')
predicted_marks = model.predict(hours)
plt.plot(hours, predicted_marks, color='red', label='Regression Line')
plt.xlabel('Study Hours')
plt.ylabel('GPA')
plt.grid(True)
plt.legend()
plt.show()
```

```python
new_hours = np.array([[5]])
predicted_gpa = model.predict(new_hours)
print(f"Predicted GPA for {new_hours[0][0]} study hours: {predicted_gpa[0][0]:.2f}") 
```

# Mean Squeare Error (MSE) Calculation

```python
# The algorithm for MSE
def mean_squared_error(y_true:list, y_pred:list) -> float:
    n = len(y_true)
    mse = sum((yt - yp) ** 2 for yt, yp in zip(y_true, y_pred)) / n
    return mse
```

# Key Insights:
1. MSE is a common metric for evaluating the performance of regression models.
2. It measures the average squared difference between actual and predicted values.
3. A lower MSE indicates a better fit of the model to the data.
4. MSE is sensitive to outliers, as it squares the errors, giving more weight to larger errors.
5. MSE can be used to compare different models or to tune model parameters for better performance.

# What is COST FUNCTION?
Cost function is a mathematical function that measures the error or difference between predicted values and actual values in a machine learning model. It quantifies how well the model's predictions match the true outcomes, and the goal of training a model is to minimize this cost function to improve the model's accuracy. In regression problems, common cost functions include Mean Squared Error (MSE) and Mean Absolute Error (MAE). In classification problems, cost functions like Cross-Entropy Loss are often used. The choice of cost function depends on the type of problem being solved and the specific requirements of the model.

# Gradient Descent 
Gradient Descent is an optimization algorithm used to minimize the cost function in machine learning models. It works by iteratively adjusting the model's parameters (weights) in the direction of the steepest descent of the cost function. The algorithm calculates the gradient (partial derivatives) of the cost function with respect to each parameter and updates the parameters by moving them in the opposite direction of the gradient, scaled by a learning rate. This process continues until convergence, where the cost function reaches a minimum value, indicating that the model has been trained effectively. Gradient Descent can be applied to various types of models, including linear regression, logistic regression, and neural networks.
